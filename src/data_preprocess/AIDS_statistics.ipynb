```python
#####**************** put this file at the same file with .gexf files, and then run *********##########

import networkx as nx
import numpy as np
import os
```

```python
from networkx import release
print(release.version)
```

```python
count = 0
num_nodes = {}
for filename in os.listdir("../AIDS/"):
    if filename.endswith(".gexf"):
        f = '../AIDS/'+filename
        G = nx.read_gexf(f)
        num_nodes[filename] = len(G) 
        count +=1
print(count)
print(num_nodes)
```

```python
num_list = list(num_nodes.values())
print(len(num_list))
print(num_list)
```

```python
max_size = max(num_list)
print(max_size)
max_size_enzymes = [k for k,v in num_nodes.items() if v == max_size]
print(max_size_enzymes)
```

```python
min_size = min(num_list)
print(min_size)
min_size_enzymes = [k for k,v in num_nodes.items() if v == min_size]
print(min_size_enzymes)
```

```python
print(np.std(num_list))
```

```python
print(np.mean(num_list))
```

```python
size=list(range(0,max_size+1))
print(size)
freq = [0]*(max_size+1)
for num in num_list:
    freq[num] += 1
print(freq)
```

```python
import matplotlib.pyplot as plt
plt.plot(size,freq , 'ro-',markerfacecolor='white',label="AIDS",linewidth=0.5, markersize=5)
plt.xlabel("#nodes")
plt.ylabel("#graphs")
plt.title('Graph Size Distribution')
plt.legend(loc='best')
plt.grid(linestyle='-', color='grey')
plt.savefig('../AIDS/graph_size_distribution.eps', format='eps')
```

```python
import sys
from bs4 import BeautifulSoup as Soup
import numpy as np
import os

node_label_dict = {}
edge_label_dict = {}
count = 0
for filename in os.listdir("../dataset/AIDS/"):
    if filename.endswith(".gexf"):
        f = '../dataset/AIDS/'+filename
        handler = open(f).read()
        soup = Soup(handler, "xml")
        for node in soup.findAll('node'):
            node_attvalues = node.findAll('attvalues')[0]
            node_attvalue = node_attvalues.findAll('attvalue')[0]
            node_attrs = dict(node_attvalue.attrs)
            node_label = node_attrs['value']
            if node_label in node_label_dict:
                node_label_dict[node_label] += 1
            else:
                node_label_dict[node_label] = 1
        for edge in soup.findAll('edge'):
            edge_attvalues = edge.findAll('attvalues')[0]
            edge_attvalue = edge_attvalues.findAll('attvalue')[0]
            edge_attrs = dict(edge_attvalue.attrs)
            edge_label = edge_attrs['value']
            if edge_label in edge_label_dict:
                edge_label_dict[edge_label] += 1
            else:
                edge_label_dict[edge_label] = 1
        count +=1
print(node_label_dict)
print(edge_label_dict)
```

```python
import csv
with open('../dataset/AIDS/node_dict.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    for key, value in node_label_dict.items():
        writer.writerow([key, value])
```

```python
import csv
with open('../dataset/AIDS/edge_dict.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    for key, value in edge_label_dict.items():
        writer.writerow([key, value])
```
